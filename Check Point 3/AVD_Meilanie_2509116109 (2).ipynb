```python
print("="*60)
print("CEK INCONSISTENT VALUES - KOLOM NUMERICAL")
print("="*60)

for col in numerical_cols:
    print(f"\n{col.upper()}:")
    print(f"  Min: {df[col].min()}")
    print(f"  Max: {df[col].max()}")

    if df[col].min() < 0:
        print(f"  WARNING: Terdapat nilai negatif!")
        print(f"  Jumlah nilai negatif: {(df[col] < 0).sum()}")

    zero_count = (df[col] == 0).sum()
    if zero_count > 0:
        print(f"  Jumlah nilai 0: {zero_count} ({zero_count/len(df)*100:.2f}%)")

print("="*60)
print("CEK NILAI 'UNKNOWN' ATAU MISSING YANG TERSAMAR")
print("="*60)

unknown_keywords = ['unknown', 'Unknown', 'UNKNOWN', 'n/a', 'N/A', 'na', 'NA', '-', '?']

for col in categorical_cols:
    unknown_count = 0
    found_unknowns = []

    for keyword in unknown_keywords:
        count = (df[col] == keyword).sum()
        if count > 0:
            unknown_count += count
            found_unknowns.append(f"{keyword}: {count}")

    if unknown_count > 0:
        print(f"\n{col.upper()}:")
        print(f"  Total 'unknown' values: {unknown_count} ({unknown_count/len(df)*100:.2f}%)")
        print(f"  Detail: {', '.join(found_unknowns)}")
```

**Insight - Inconsistent Values:**

3. Nilai negatif: Pada kolom yang seharusnya positif (seperti usia, saldo), adanya nilai negatif mengindikasikan error data
4. Nilai nol yang mencurigakan: Banyaknya nilai 0 pada kolom tertentu perlu diperiksa apakah wajar atau merupakan missing value
5. Typo atau variasi penulisan: Kategori yang sama tapi ditulis berbeda (misal: 'single' vs 'Single')

**Tindakan yang perlu dilakukan:**
- Standardisasi format penulisan

- Validasi dan perbaiki nilai yang tidak logis

### **2.7 Cek Outliers**

```python
print("="*60)
print("DETEKSI OUTLIERS - METODE IQR")
print("="*60)

outliers_summary = []

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_count = len(outliers)
    outlier_percentage = (outlier_count / len(df)) * 100

    outliers_summary.append({
        'Kolom': col,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outlier Count': outlier_count,
        'Outlier %': round(outlier_percentage, 2)
    })

outliers_df = pd.DataFrame(outliers_summary)
print(outliers_df.to_string(index=False))
```

```python
fig, axes = plt.subplots(nrows=(len(numerical_cols)+2)//3, ncols=3, figsize=(15, 4*((len(numerical_cols)+2)//3)))
axes = axes.flatten()

for idx, col in enumerate(numerical_cols):
    sns.boxplot(data=df, y=col, ax=axes[idx], color='lightblue')
    axes[idx].set_title(f'Boxplot: {col}', fontsize=12, fontweight='bold')
    axes[idx].set_ylabel(col, fontsize=10)

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers_count = len(df[(df[col] < Q1-1.5*IQR) | (df[col] > Q3+1.5*IQR)])
    axes[idx].text(0.5, 0.98, f'Outliers: {outliers_count}',
                   transform=axes[idx].transAxes,
                   ha='center', va='top',
                   bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.5))

for idx in range(len(numerical_cols), len(axes)):
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy import stats

print("\n" + "="*60)
print("DETEKSI OUTLIERS - METODE Z-SCORE")
print("="*60)

zscore_outliers = []

for col in numerical_cols:
    z_scores = np.abs(stats.zscore(df[col]))
    outliers = df[z_scores > 3]  # Threshold Z-score > 3
    outlier_count = len(outliers)
    outlier_percentage = (outlier_count / len(df)) * 100

    zscore_outliers.append({
        'Kolom': col,
        'Outlier Count (Z>3)': outlier_count,
        'Outlier %': round(outlier_percentage, 2)
    })

zscore_df = pd.DataFrame(zscore_outliers)
print(zscore_df.to_string(index=False))
```

**Insight - Outliers:**

Berdasarkan deteksi outliers:

1. **Metode IQR (Interquartile Range)**:
   - Mendeteksi nilai yang berada di luar 1.5 * IQR dari Q1 dan Q3
   - Lebih sensitif terhadap outliers
   - Cocok untuk distribusi yang tidak terlalu skewed

2. **Metode Z-Score**:
   - Mendeteksi nilai yang memiliki Z-score > 3 (atau < -3)
   - Mengasumsikan distribusi normal
   - Lebih robust untuk dataset besar

3. **Interpretasi Outliers**:
   - Outliers pada usia: Mungkin ada nasabah yang sangat muda atau sangat tua
   - Outliers pada saldo: Nasabah dengan saldo sangat tinggi atau sangat rendah (bahkan negatif)
   - Outliers pada durasi panggilan: Panggilan yang sangat singkat atau sangat lama
   - Outliers pada jumlah kontak: Nasabah yang dihubungi berulang kali

4. **Penanganan Outliers**:
   - Jangan langsung dihapus - outliers bisa mengandung informasi penting
   - Untuk business case ini, outliers mungkin merepresentasikan segment nasabah tertentu
   - Pertimbangkan:
     * Winsorization (cap nilai ekstrem)
     * Transformasi log untuk mengurangi skewness
     * Binning untuk mengelompokkan nilai ekstrem
     * Biarkan jika outliers adalah natural (misalnya: nasabah VIP dengan saldo tinggi)

5. **Dampak pada Model**:
   - Outliers dapat mempengaruhi algoritma yang sensitive terhadap scale (Linear Regression, KNN)
   - Tree-based models (Decision Tree, Random Forest) lebih robust terhadap outliers

**Insight - Distribution:**

Dari analisis distribusi data:

1. **Distribusi Normal vs Skewed**:
   - Right-skewed (positively skewed): Data menumpuk di kiri, tail panjang ke kanan
     * Contoh: Saldo rekening - kebanyakan nasabah punya saldo rendah, sedikit yang tinggi
     * Implikasi: Mean > Median
   - Left-skewed (negatively skewed): Data menumpuk di kanan, tail panjang ke kiri
     * Implikasi: Mean < Median
   - Normal distribution: Data tersebar merata di sekitar mean
     * Implikasi: Mean â‰ˆ Median

2. **Skewness Interpretation**:
   - |Skewness| < 0.5: Fairly symmetrical (mendekati normal)
   - 0.5 < |Skewness| < 1: Moderately skewed
   - |Skewness| > 1: Highly skewed

3. **Kurtosis Interpretation**:
   - Kurtosis â‰ˆ 3: Normal distribution
   - Kurtosis > 3: Heavy-tailed (banyak outliers)
   - Kurtosis < 3: Light-tailed (sedikit outliers)

4. **Implikasi untuk Modeling**:
   - Data yang highly skewed mungkin perlu transformasi (log, sqrt, box-cox)
   - Algoritma seperti Linear Regression mengasumsikan normalitas
   - Tree-based models lebih toleran terhadap non-normal distribution

## **4. DATA PREPARATION**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load dataset
path = "/content/sample_data/bank.csv"
df = pd.read_csv(path)
df_clean = df.copy()

print("Dataset awal berhasil dimuat.")
print(f"Jumlah baris: {df_clean.shape[0]}, Jumlah kolom: {df_clean.shape[1]}")
```

###**BAGIAN 1: PENANGANAN DATA**

### **4.1 Tipe Data**

```python
print("\n--- 1.1 PENANGANAN TIPE DATA ---")

month_map = {
    'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4,
    'may': 5, 'jun': 6, 'jul': 7, 'aug': 8,
    'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
}
df_clean['month'] = df_clean['month'].map(month_map)
print("Kolom 'month' dikonversi dari string ke integer (1-12).")
```

#### Penjelasan:
Kolom month menyimpan nama bulan dalam format string (jan, feb, dst.) yang seharusnya bersifat ordinal atau berurutan. Kolom day sudah bertipe integer dan tidak perlu diubah.

### **4.2 Inconsistent Values**

```python
print("\n--- 1.2 PENANGANAN INCONSISTENT VALUES ---")

# Job: drop baris 'unknown' (hanya 0.63%)
before = len(df_clean)
df_clean = df_clean[df_clean['job'] != 'unknown']
print(f"'job' unknown: {before - len(df_clean)} baris di-drop.")

# Education: imputasi dengan modus
modus_education = df_clean['education'].mode()[0]
df_clean['education'] = df_clean['education'].replace('unknown', modus_education)
print(f"'education' unknown: diimputasi dengan modus = '{modus_education}'.")

# Contact: ganti 'unknown' jadi 'not_specified'
df_clean['contact'] = df_clean['contact'].replace('unknown', 'not_specified')
print("'contact' unknown: diganti menjadi 'not_specified'.")

# Poutcome: ganti 'unknown' jadi 'no_previous'
df_clean['poutcome'] = df_clean['poutcome'].replace('unknown', 'no_previous')
print("'poutcome' unknown: diganti menjadi 'no_previous'.")

# Pdays: -1 artinya tidak pernah dihubungi, buat flag binary
df_clean['was_contacted_before'] = (df_clean['pdays'] != -1).astype(int)
df_clean['pdays'] = df_clean['pdays'].replace(-1, np.nan)
print("'pdays': nilai -1 dikonversi ke NaN, kolom 'was_contacted_before' dibuat.")

print(f"\nUkuran dataset: {df_clean.shape}")
```

#### Penjelasan:
Beberapa kolom mengandung nilai unknown yang merupakan missing value tersamar. Strategi penanganan disesuaikan dengan jumlah dan konteks bisnis masing-masing kolom:

- *job* (0.63%): Baris dengan nilai unknown di-drop karena jumlahnya sangat kecil sehingga tidak berdampak signifikan pada data.
- *education* (4.45%): Diimputasi dengan modus (nilai terbanyak = secondary) karena persentasenya masih kecil dan nilai terbanyak dapat mewakili data yang hilang.
- *contact* (21.02%): Diganti menjadi not_specified karena jumlahnya terlalu besar untuk di-drop maupun diimputasi secara sembarangan.
- *poutcome* (74.59%): Diganti menjadi no_previous karena berdasarkan konteks bisnis, nilai unknown pada kolom ini berarti nasabah memang belum pernah dihubungi sebelumnya, bukan data yang hilang secara acak.

Selain itu, kolom pdays menggunakan nilai -1 sebagai kode khusus yang berarti nasabah belum pernah dihubungi. Nilai -1 bukan nilai hari yang valid, sehingga dikonversi menjadi NaN dan dibuat kolom baru was_contacted_before sebagai penanda binary (0/1).


### **4.3 Missing Values**

```python
print("\n--- 1.3 PENANGANAN MISSING VALUES ---")

missing = df_clean.isnull().sum()
missing = missing[missing > 0]

if len(missing) > 0:
    print("Missing values ditemukan:")
    print(missing)
    median_pdays = df_clean[df_clean['was_contacted_before'] == 1]['pdays'].median()
    df_clean['pdays'] = df_clean['pdays'].fillna(median_pdays)
    print(f"'pdays' NaN diimputasi dengan median = {median_pdays:.1f} hari.")
else:
    print("Tidak ada missing values.")

print(f"Total missing values: {df_clean.isnull().sum().sum()}")
```

#### Penjelasan:
Setelah konversi nilai -1 pada pdays menjadi NaN, muncul missing values pada kolom tersebut. Imputasi dilakukan menggunakan median dari nasabah yang pernah dihubungi sebelumnya (was_contacted_before = 1). Median dipilih karena distribusi pdays cenderung right-skewed, sehingga median lebih representatif dibandingkan mean yang mudah terpengaruh oleh nilai ekstrem.

### **4.4 Duplicated Values**

```python
print("\n--- 1.4 PENANGANAN DUPLICATED VALUES ---")

dup_count = df_clean.duplicated().sum()
print(f"Jumlah baris duplikat: {dup_count}")
if dup_count > 0:
    df_clean = df_clean.drop_duplicates(keep='first')
    print(f"{dup_count} baris duplikat dihapus. Ukuran dataset: {df_clean.shape}")
else:
    print("Tidak ada duplikat.")
```

#### Penjelasan:
Pemeriksaan duplikat dilakukan setelah seluruh proses transformasi sebelumnya untuk memastikan tidak ada baris yang identik akibat proses pembersihan. Hasilnya tidak ditemukan baris duplikat pada dataset ini.

### **4.5 Outliers**

```python
print("\n--- 1.5 PENANGANAN OUTLIERS ---")

cols_to_winsorize = ['balance', 'duration', 'campaign', 'previous']

for col in cols_to_winsorize:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    before_count = ((df_clean[col] < lower) | (df_clean[col] > upper)).sum()
    df_clean[col] = df_clean[col].clip(lower=lower, upper=upper)
    print(f"'{col}': {before_count} outlier di-cap ke [{lower:.1f}, {upper:.1f}]")

print(f"\nUkuran dataset setelah outlier handling: {df_clean.shape}")
```

#### Penjelasan:
Metode yang digunakan adalah Winsorization (Capping) berdasarkan batas IQR. Nilai di bawah Q1 - 1.5Ã—IQR dicap ke lower bound dan nilai di atas Q3 + 1.5Ã—IQR dicap ke upper bound. Metode ini dipilih karena menghapus outlier akan menghilangkan terlalu banyak data (misalnya previous memiliki 11.27% outlier). Selain itu, beberapa outlier memiliki makna bisnis yang penting, seperti nasabah dengan saldo sangat tinggi atau durasi panggilan sangat lama yang justru menandakan tingkat keterlibatan tinggi. Kolom yang di-winsorisasi adalah balance, duration, campaign, dan previous. Kolom age tidak diubah karena outlier-nya hanya 1.53% dan rentang usia 18-95 masih tergolong wajar.

Nilai `unknown` pada beberapa kolom ditangani dengan strategi yang berbeda-beda sesuai persentase dan konteks bisnisnya, mulai dari drop baris pada kolom `job`, imputasi modus pada kolom `education`, hingga penggantian label pada kolom `contact` dan `poutcome`.
