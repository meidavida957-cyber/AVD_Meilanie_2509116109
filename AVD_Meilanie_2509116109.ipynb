## **1. BUSINESS UNDERSTANDING**

Business Understanding merupakan tahap pertama dalam metodologi CRISP-DM (Cross-Industry Standard Process for Data Mining). Pada tahap ini, kita akan memahami konteks bisnis, mengidentifikasi masalah, dan menentukan tujuan analisis data.

### **1.1 Business Objective**

#### **Latar Belakang**

Dataset Bank Marketing ini berisi data dari kampanye pemasaran langsung (direct marketing) sebuah institusi perbankan di Portugal. Kampanye pemasaran ini dilakukan melalui telepon untuk menawarkan produk deposito berjangka (term deposit) kepada nasabah.

#### **Penjelasan Data**

Dataset ini mencakup informasi tentang:
- **Data Demografis Nasabah**: usia, pekerjaan, status pernikahan, pendidikan
- **Data Finansial**: status kredit, saldo rekening, pinjaman
- **Data Kampanye**: jumlah kontak, durasi panggilan, hasil kampanye sebelumnya
- **Data Ekonomi**: indikator ekonomi makro saat kampanye berlangsung
- **Target Variable**: Apakah nasabah berlangganan deposito berjangka atau tidak (yes/no)

#### **Identifikasi Masalah**

Permasalahan utama yang dihadapi oleh bank adalah:

1. **Efisiensi Rendah dalam Kampanye Marketing**
   - Banyak waktu dan sumber daya terbuang untuk menghubungi nasabah yang tidak berminat
   - Conversion rate (tingkat keberhasilan) kampanye marketing masih rendah
   - Biaya operasional tinggi karena pendekatan yang tidak tertarget

2. **Kurangnya Pemahaman tentang Profil Nasabah Potensial**
   - Tidak ada strategi yang jelas dalam menentukan target nasabah
   - Sulit mengidentifikasi karakteristik nasabah yang cenderung tertarik pada produk deposito

3. **Optimalisasi ROI (Return on Investment)**
   - Perlu meningkatkan efektivitas kampanye dengan budget yang terbatas
   - Memaksimalkan hasil dengan meminimalkan usaha yang sia-sia

#### **Tujuan Analisis**

Berdasarkan permasalahan di atas, tujuan dari analisis ini adalah:

1. **Mengidentifikasi pola dan karakteristik** nasabah yang berpotensi tinggi untuk berlangganan deposito berjangka

2. **Membangun model prediksi** yang dapat memprediksi nasabah mana yang kemungkinan besar akan tertarik dengan produk deposito

3. **Meningkatkan efisiensi kampanye marketing** dengan fokus pada nasabah yang tepat, sehingga:
   - Mengurangi biaya operasional
   - Meningkatkan conversion rate
   - Mengoptimalkan penggunaan waktu tim marketing

4. **Memberikan insight bisnis** yang actionable untuk strategi marketing yang lebih efektif di masa depan

#### **Manfaat untuk Bisnis**

- **Penghematan Biaya**: Mengurangi jumlah panggilan yang tidak produktif
- **Peningkatan Revenue**: Fokus pada nasabah dengan probabilitas tinggi meningkatkan penjualan deposito
- **Efisiensi Operasional**: Tim marketing dapat bekerja lebih efektif dan efisien
- **Customer Experience**: Nasabah yang tidak berminat tidak terganggu dengan panggilan yang tidak relevan

### **1.2 Assess Situation**

Pada tahap ini, kita akan mengevaluasi kondisi dataset dan mengidentifikasi tantangan yang mungkin dihadapi dalam analisis.

#### **Kondisi Dataset**

Berdasarkan eksplorasi awal dataset Bank Marketing:

1. **Ukuran Dataset**
   - Dataset memiliki ribuan record data nasabah
   - Jumlah fitur/kolom yang cukup komprehensif untuk analisis

2. **Jenis Data**
   - Kombinasi data numerical (usia, saldo, durasi panggilan, dll)
   - Data categorical (pekerjaan, pendidikan, status pernikahan, dll)
   - Target variable bersifat binary (yes/no)

#### **Tantangan yang Dihadapi**

1. **Class Imbalance**
   - Kemungkinan terdapat ketidakseimbangan antara nasabah yang subscribe vs tidak subscribe
   - Jumlah nasabah yang berlangganan deposito biasanya lebih sedikit dibanding yang tidak
   - Hal ini dapat mempengaruhi performa model prediksi

2. **Missing Values**
   - Perlu pengecekan apakah ada data yang hilang atau tidak lengkap
   - Beberapa kolom mungkin memiliki nilai 'unknown' yang perlu ditangani

3. **Outliers**
   - Data seperti durasi panggilan, saldo rekening, atau usia mungkin memiliki outliers
   - Perlu identifikasi dan penanganan yang tepat agar tidak mengganggu analisis

4. **Multicollinearity**
   - Beberapa variabel mungkin saling berkorelasi tinggi
   - Perlu identifikasi untuk menghindari redundansi dalam model

5. **Feature Engineering**
   - Data categorical perlu di-encode menjadi numerical
   - Mungkin perlu pembuatan fitur baru dari kombinasi fitur yang ada
   - Normalisasi/standarisasi untuk fitur numerical dengan skala berbeda

6. **Data Quality**
   - Perlu validasi konsistensi data
   - Pengecekan data duplikat
   - Verifikasi kelogisan data (misalnya: usia tidak negatif)

#### **Resources yang Tersedia**

- Dataset lengkap dengan berbagai fitur demografis, finansial, dan kampanye
- Tools: Python dengan library data science (Pandas)
- Platform: Google Colab untuk komputasi

#### **Risiko dan Asumsi**

**Risiko:**
- Model mungkin bias terhadap kelas mayoritas karena imbalanced data
- Overfitting jika model terlalu kompleks
- Hasil prediksi mungkin tidak 100% akurat

**Asumsi:**
- Data yang dikumpulkan representatif terhadap populasi nasabah
- Pola perilaku nasabah di masa lalu akan relevan untuk prediksi masa depan
- Kondisi ekonomi dan pasar relatif stabil

### **3.3 Analytic Goals Project Plan**

Berdasarkan business objective dan kondisi dataset, berikut adalah goals analitik yang akan dicapai:

#### **Goals Utama**

1. **Exploratory Data Analysis (EDA)**
   - Memahami distribusi setiap variabel dalam dataset
   - Mengidentifikasi pola dan tren dalam data
   - Menemukan hubungan antara variabel independen dengan target variable
   - Visualisasi data untuk mendapatkan insight

2. **Data Preprocessing**
   - Menangani missing values dengan metode yang sesuai
   - Deteksi dan treatment outliers
   - Encoding variabel categorical
   - Feature scaling/normalization
   - Handling imbalanced data (SMOTE, undersampling, atau class weights)

3. **Feature Selection & Engineering**
   - Identifikasi fitur-fitur yang paling berpengaruh terhadap keputusan subscribe
   - Eliminasi fitur yang tidak relevan atau redundant
   - Membuat fitur baru jika diperlukan untuk meningkatkan performa model
   - Analisis korelasi antar fitur

4. **Model Development**
   - Membangun model machine learning untuk prediksi subscription
   - Mencoba berbagai algoritma (Logistic Regression, Decision Tree, Random Forest, XGBoost, dll)
   - Hyperparameter tuning untuk optimasi model
   - Cross-validation untuk validasi performa model

5. **Model Evaluation**
   - Evaluasi model dengan metrik yang sesuai:
     - Accuracy
     - Precision (penting untuk minimalisir False Positive)
     - Recall (penting untuk menangkap sebanyak mungkin nasabah potensial)
     - F1-Score
     - ROC-AUC Score
   - Confusion Matrix untuk memahami prediksi model
   - Feature importance analysis

6. **Business Insights**
   - Mengidentifikasi profil nasabah yang paling berpotensi subscribe
   - Memberikan rekomendasi strategi marketing berdasarkan hasil analisis
   - Estimasi peningkatan ROI dengan menggunakan model prediksi

#### **Kriteria Kesuksesan**

Analisis dianggap berhasil jika:

1. **Model Performance**
   - ROC-AUC Score minimal 0.75 (baik)
   - Precision minimal 60% untuk meminimalkan waktu terbuang pada nasabah yang tidak berminat
   - Recall minimal 70% untuk menangkap sebagian besar nasabah potensial

2. **Business Impact**
   - Dapat mengidentifikasi minimal 3-5 faktor utama yang mempengaruhi keputusan subscribe
   - Memberikan segmentasi nasabah yang jelas (high, medium, low potential)
   - Proyeksi peningkatan conversion rate minimal 20-30% dengan targeting yang lebih baik

3. **Actionable Insights**
   - Rekomendasi yang praktis dan dapat diimplementasikan oleh tim marketing
   - Visualisasi yang mudah dipahami oleh stakeholder non-teknis

#### **Deliverables**

1. Model prediksi yang dapat digunakan untuk scoring nasabah
2. Dashboard/visualisasi untuk memahami karakteristik nasabah potensial
3. Laporan analisis dengan insight dan rekomendasi bisnis
4. Documentation lengkap untuk reproduksi dan deployment model
